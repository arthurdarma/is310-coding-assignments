```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url='https://humanist.kdl.kcl.ac.uk/Archives/'
response = requests.get(url)
soup = BeautifulSoup(response.text, features="html.parser")


mainlinks = soup.find_all('a')[1:]
mainlinks
```

```python
convertedurl = url + mainlinks[0].get('href')
sub_response = requests.get(convertedurl)
sub_soup = BeautifulSoup(sub_response.text, "html.parser")
link1 = sub_soup.find_all('a')[1:-1]
link1 
```

```python
link1[0].get_text().split(".")[1]
```

```python
with open('convertedtext.csv', 'w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Link', 'Text','Years'])
    for link in link1:
        volume_url = convertedurl + link.get('href')
        volume_response = requests.get(volume_url)
        volume_soup = BeautifulSoup(volume_response.text, "html.parser")
        writer.writerow([link.get_text(), volume_soup.get_text()[:10000],link.get_text().split(".")[1]])

convertedtext = pd.read_csv('convertedtext.csv')
convertedtext

```

```python
currenturl = url + mainlinks[1].get('href')
sub_response = requests.get(currenturl)
sub_soup = BeautifulSoup(sub_response.text, "html.parser")
link2 = sub_soup.find_all('a')[1:]
link2 
```

```python
with open('currenttext.csv', 'w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Link', 'Text','Volume'])
    for link in link2:
        volume_url = currenturl + link.get('href')
        volume_response = requests.get(volume_url)
        volume_soup = BeautifulSoup(volume_response.text, "html.parser")
        writer.writerow([link.get_text(), volume_soup.get_text()[:10000],link.get_text().split(".")[1]])

currenttext = pd.read_csv('currenttext.csv')
currenttext
```
